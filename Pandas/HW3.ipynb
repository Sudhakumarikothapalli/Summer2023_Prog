### Task 17

Alice, Mark, Bill, Tony, Jason, Ravi, Sandeep, Mounika, John and Curtis took three graduate courses `CS 5000`, `CS 5110` and `CS 5600` in Spring 2022. To simulate the midterm scores they got for the courses, please use a random seed 123 to generate a $10 \times 3$ DataFrame called scores to store the 10 stduents' scores. The scores are random integers in the range of `[30, 100]`

```python
np.random.seed(123)
mid_term_report = pd.DataFrame(np.random.randint(30,101,size=(10, 3)), columns=['CS 5000','CS 5110','CS 5600'],index=['Alice', 'Mark', 'Bill', 'Tony', 'Jason', 'Ravi', 'Sandeep', 'Mounika', 'John', 'Curtis'])
mid_term_report
```

### Task 18

Repeat Task 19 to generate the final scores for those stduents. Please use a random seed of 456 to generate the data

```python
np.random.seed(456)
final_report = pd.DataFrame(np.random.randint(30,101,size=(10, 3)), columns=['CS 5000','CS 5110','CS 5600'],index=['Alice', 'Mark', 'Bill', 'Tony', 'Jason', 'Ravi', 'Sandeep', 'Mounika', 'John', 'Curtis'])
final_report
```

### Task 19
Concatenate the two DataFrames from Task 19 and Task 20 into a new DataFrame with multi index

```python
complete_report=pd.concat([mid_term_report, final_report],keys= ['Midterm', 'Final'],axis=0)
complete_report
```

### Task 20

### Task 21
